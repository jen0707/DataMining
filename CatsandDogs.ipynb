```python
import kagglehub

path = kagglehub.dataset_download("scarb7/catsanddogs-dummy")

print("Path to dataset files:", path)
```

```python
import os

files = os.listdir(path)
print(files)
```

```python
import pandas as pd

csv_path = os.path.join(path, "CatsAndDogs.csv")
df = pd.read_csv(csv_path)
df.head()
```

```python
cats = df[df["Animal"] == 0]
dogs = df[df["Animal"] == 1]
```

```python
print(cats)
print(dogs)
```

```python
import matplotlib.pyplot as plt

plt.scatter(cats["Height"], cats["Weight"], c='blue', label='Cat')
plt.scatter(dogs["Height"], dogs["Weight"], c='red', label='Dog')
plt.xlabel("Height")
plt.ylabel("Weight")
plt.legend()
plt.title("Height vs Weight")
plt.show()
```

```python
X = df.drop("Animal", axis=1)  # 특성
Y = df["Animal"] # 0 = cat, 1 = dog
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(X, Y)
```

```python
kn.score(X, Y)
```
